```python
# Import dependencies
import pandas as pd
import pandas.io.sql as psql
import matplotlib.pyplot as plt
# Import the SQL database into Pand
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:selin@localhost:5432/employeeDB')
connection = engine.connect()
```

```python
# Create dataframe for each tables
departments=psql.read_sql("SELECT * FROM departments", connection)
dept_emp=psql.read_sql("SELECT * FROM dept_emp", connection)
dept_manager=psql.read_sql("SELECT * FROM dept_manager", connection)
employees=psql.read_sql("SELECT * FROM employees", connection)
salaries=psql.read_sql("SELECT * FROM salaries", connection)
titles=psql.read_sql("SELECT * FROM titles", connection)
```

```python
# Create bins and labels to create salary ranges
bins=[39000,49000,59000,69000,79000,89000,99000,109000,
      119000,129000,139000]
labels=["40k-50k","50k-60k","60k-70k","70k-80k",
        "80k-90k","90k-100k","100k-110k","110k-120k",
        "120k-130k","130k-140k"]
salaries["Salary Range"]=pd.cut(salaries["salary"], bins,labels=labels)
```

```python
# Ceate a histogram to visualize the most common salary ranges for employees.

plt.figure(figsize=(10,6))
salaryRangePlot=plt.hist(salaries["salary"], bins=bins, 
                         color="purple")
plt.xlabel("Salary Ranges")
plt.xticks(bins, labels, rotation=45)
plt.title("Salary Range Frequency")
```

```python
# Merge salaries and titles data tables
merge_df=pd.merge(salaries, titles, on="emp_no", how='left')
```

```python
# Create a bar chart of average salary by title.
salaryTitle=pd.DataFrame(
    merge_df.groupby('title')['salary'].mean()).sort_values(
    ["salary"])
salaryTitlePlot=salaryTitle.plot(kind='bar', figsize=(10,6), color='purple')
salaryTitlePlot.set_xlabel("Employee Title")
salaryTitlePlot.set_ylabel("Avg Salary")
salaryTitlePlot.set_title("Avg Salary by Employee Titles")
```
